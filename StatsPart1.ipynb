**Measure Of Central Tendency**


```python
ages = [10,20,30,35,23,56,45,14,13,17,13,18,19,32,56,100,95,24,77,39,40]
```

```python
import numpy as np
print(np.mean(ages))
print(np.median(ages))
```

**5 Number Summary**

```python
import numpy as np
q1 = np.percentile(ages,25)
q3 = np.percentile(ages,75)
print(q1,q3)
```

```python
##To check Outlier [Lower Fence-  Higher Fence]
IQR = q3-q1
lower_fence = q1 - 1.5*(IQR)
higher_fence = q3 + 1.5*(IQR)
print(lower_fence , higher_fence)
```

**Measure Of Dispersion**

1. **Variance**
2. **Standard Deviation**

```python
statistics.variance(ages)
```

```python
np.var(ages, axis = 0)
```

```python
def variance(data , dof=0):
  n = len(data)
  # mean of the data
  mean = sum(data)/n
  # variance of the data
  deviation = [(x-mean)**2 for x in data]
  variance = sum(deviation)/ (n-1)
  return variance
  # last ma sum(deviation)/n-1 vaneko sample variance and /n vaneko chai population variance
```

```python
variance(ages,dof=1)
```

```python
# population variance
statistics.pvariance(ages)
```

```python
# Standard deviation = square rooot under variance ho hai so....
import math
math.sqrt(statistics.pvariance(ages))
```

**Check Whether the Distribution is Normally Distributed or Not**

```python
#### If you want to check whether feature is guassian or normally distributed
#### Q-Q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
## create a normal distributed data
s=np.random.normal(0.5,0.2,400)
plot_data(s)
```

```python
plot_data(df['sepal_width'])
```
