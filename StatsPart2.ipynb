```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, poisson

# ------------------------------
# PART 1: PDF and CDF (Continuous) – Height Example
# ------------------------------
# Simulate height data (normally distributed)
np.random.seed(0)
heights = np.random.normal(loc=170, scale=10, size=1000)

# Plot PDF
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(heights, kde=True, stat="density", bins=30, color='skyblue', label='PDF')
plt.title("PDF - Heights")
plt.xlabel("Height (cm)")
plt.legend()

# Plot CDF
plt.subplot(1, 2, 2)
sns.ecdfplot(heights, label='CDF', color='darkorange')
plt.title("CDF - Heights")
plt.xlabel("Height (cm)")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# PART 2: PMF (Discrete) – Email Count Example

# Simulate number of emails received in a day (Poisson distributed)
email_counts = np.random.poisson(lam=3, size=1000)

# Calculate PMF
values, counts = np.unique(email_counts, return_counts=True)
pmf = counts / len(email_counts)

# Plot PMF
plt.figure(figsize=(6, 4))
plt.stem(values, pmf, basefmt=" ",)
plt.title("PMF - Emails Received per Day")
plt.xlabel("Number of Emails")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```
